```python
# basic plot (line plot default) using matplotlib

import matplotlib.pyplot as plt
plt.plot([1,5,3,6,9,7])
plt.show()
```

```python
# plotting with axis label and legend

import matplotlib.pyplot as plt
plt.plot([1,5,3,6,9,7])
plt.plot([12,5,4,3,6,1])
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value"])
plt.show()
```

```python
# in the below we connect the two point (0,1) and (0,12) , (0,5) and (0,5) and so on

import matplotlib.pyplot as plt
plt.plot([[1,5,3,6,9,7],[12,5,4,3,6,1]])
plt.show()
```

```python
# In the below example plot we join the point (0,1) (0,12) and (0,1) and so on

import matplotlib.pyplot as plt
plt.plot([[1,5,3,6,9,7],[12,5,4,3,6,1],[1,5,3,6,9,7]])
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["x1","x2","x3","x4","x5","x6"])
plt.show()
```

```python
# In the below example we get an error because matplotlib not work on 3D plot

import matplotlib.pyplot as plt
plt.plot([[[1,5,3,6,9,7],[12,5,4,3,6,1],[1,5,3,6,9,7]]])
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])
plt.show()
```

```python
# "rh-." red line (r) , point in hexagon (h) and line type is dashed dot (-.) line 

import matplotlib.pyplot as plt
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.")
# x axis and y axis plot
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])
plt.show()
```

```python
# plotting only point (r--red and h-- hexagon point)

import matplotlib.pyplot as plt
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh")
# x axis and y axis plot
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])
# plt.plot([12,5,4,3,6,1])
plt.show()
```

```python
# plotting with linewidth and markersize

import matplotlib.pyplot as plt
plt.plot([12,5,4,3,6,1],"rh-.",linewidth=5,markersize=15)
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])
plt.show()
```

```python
# scatter plot in matplotlib with the size odf data points s=[10,20,30,40,50,60]

plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```

```python
# in scatter plot data point with color

plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100),c=range(1,7))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```

```python
# plotting with different marker size

import matplotlib.pyplot as plt
import matplotlib as mat
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],
            s=range(100,601,100),
            c=range(1,7),
           marker=mat.markers.CARETDOWNBASE)
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```

```python
# plotting two graph up and bottom using subplot

plt.subplot(2,1,1)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.")
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])

plt.subplot(2,1,2)
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100),c=range(1,7))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```

```python
# plotting two graph side by side using subplot

plt.subplot(1,2,1)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.")
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])

plt.subplot(1,2,2)
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100),c=range(1,7))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```

```python
# four graph plot at once using subplot

plt.subplot(2,2,1)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.")
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])

plt.subplot(2,2,2)
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100),c=range(1,7))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")

plt.subplot(2,2,3)
plt.plot([1,5,3,6,9,7],[12,5,4,3,6,1],"rh-.")
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.legend(["Age Line","Random Value","x1","x2","x3","x4"])

plt.subplot(2,2,4)
plt.scatter([1,5,3,6,9,7],[12,5,4,3,6,1],s=range(100,601,100),c=range(1,7))
plt.xlabel("index")
plt.ylabel("Age")
plt.title("index and age relation")
plt.show()
```
