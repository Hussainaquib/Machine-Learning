```python
# import all important required library and algo/model

import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

####  Data Gathering

```python
wine_data=pd.read_csv("C:/Users/91975/Downloads/archive (8)/winequality-red.csv")
print(wine_data.columns)
print()
print(wine_data.info())
```

#### Data Pre processing

```python
# split the input and output data

X,Y=wine_data.drop(["quality"],axis=1),wine_data["quality"]
```

```python
# categories X and Y as x_train,x_test,y_train,y_test with train_size=0.7

x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.7,shuffle=False)
print(x_test.shape,x_train.shape)
```

#### model selection

```python
dt=DecisionTreeClassifier(criterion="log_loss",max_features="log2")
```

#### training the model

```python
dt.fit(x_train,y_train)
```

```python
# calculating training accuracy

print(dt.score(x_train,y_train))
```

#### Testing our model

```python
# predict the x_test value

predict=dt.predict(x_test)
print(predict)
```

```python
# calculating testing accuracy

print(accuracy_score(y_test,predict))
```

```python
# printing confusion matrix

print(confusion_matrix(y_test,predict))
```
