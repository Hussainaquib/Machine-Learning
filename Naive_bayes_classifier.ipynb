```python
# importing all the required library

import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
```

```python
# (1). data Gathering

data=pd.read_csv("C:/Users/91975/Downloads/archive (7)/Naive-Bayes-Classification-Data.csv")
print(data)
```

```python
#detail information about data

print(data.info())
```

```python
# (2). Data pre processing

#seperate the input and output column
X,Y=data.drop("diabetes",axis=1),data['diabetes']
print(X,Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8)
```

```python
# (3). choose an algo

nb=GaussianNB()
```

```python
# (4). train the model

nb.fit(x_train,y_train)
print("training accuracy",nb.score(x_train,y_train))
```

```python
# (5). test the model

predicted_data=nb.predict(x_test)
print(predicted_data)
print("testing accuracy",accuracy_score(y_test,predicted_data))
#accuracy_score(actual_data,predicted_data)
```

```python
print("confusion matrix",confusion_matrix(y_test,predicted_data))
# main diagonal is right prediction and are else wrong prdiction
```
