```python
# import all required library and algo/model

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

### 1. Data gathering

```python
data=pd.read_csv("creditcard (1).csv")
data.info()
# class is output data and other column is input data
```

### 2. Data pre processing

```python
X, Y=data.drop(['Class'],axis=1),data['Class']
print(list(Y).count(0))
print(list(Y).count(1))
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.7)
```

### 3. Model Selection

```python
log_reg = LogisticRegression(max_iter=150000)
```

### 4. Train our model

```python
log_reg.fit(x_train,y_train)
```

### 5. Test our model

```python
pred = log_reg.predict(x_test)
print(pred)
print(len(pred))
```

```python
# calculating log_reg score and accuracy score

print(log_reg.score(x_train,y_train))
print(accuracy_score(y_test,pred))
```

```python
# plotting the x_test data

pca=PCA(2)
x=pca.fit(x_test).transform(x_test)
plt.scatter(x[:,0],x[:,1],c=pred)
plt.show()
```
